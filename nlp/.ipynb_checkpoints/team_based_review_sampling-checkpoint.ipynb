```python
# 한글만 추출 + 2개이상 공백 1개로 변환
def extract_korean(text):
    # 한글만 추출
    text = re.sub(r'[^ㄱ-ㅎㅏ-ㅣ가-힣\s]','',text)
    # 공백 2개 이상인경우 1개로 변환
    text = re.sub(r'\s+',' ',text)
    return text
```

```python
# 시리즈에서 중복을 제거한 형용사 추출후 리스트로 반환
def extract_adjectives_from_series(pecab, series):

    adjectives = set()  # 중복을 제거하기 위해 집합 사용
    # 시리즈의 각 요소에 대해 반복
    for text in series:
        
        pecab_pos = pecab.pos(text)
        join_text = []
        for text,pos in pecab_pos:
            if pos == 'VA':
                join_text.append(text)
        
        adjectives.update(join_text)
    
    
    return list(adjectives)
```

```python
df_1t['review2'] = extract_korean_from_series(df_1t['review'])
df_1t.head(2)
```
