```python
import pandas as pd
import numpy as np
```

```python
# 1   (하) CSV 파일을 읽고 데이터프레임으로 로드하세요.
df = pd.read_csv('인사정보.csv', encoding='utf-8')
display(df)
```

```python
# 2   (하) 데이터프레임의 처음 5행을 출력하세요.
display(df.head(5))
```

```python
# 3   (하) 데이터프레임의 구조(행과 열의 수)를 확인하세요.
print(df.shape)
```

```python
# 4   (중) 각 부서별 직원 수를 계산하세요.
print(df['부서'].value_counts())
```

```python
# 5   (중) 직급별 평균 연봉을 계산하세요.
print(df.groupby('직급')['연봉(만원)'].mean())
```

```python
# 6   (중) 2010년 이후 입사한 직원의 수를 계산하세요.
display(df[df['입사년도'] >= 2010])
```

```python
# 7   (중) 연봉이 6000만원 이상인 직원들의 정보를 출력하세요.
display(df[df['연봉(만원)'] >= 6000])
```

```python
# 8   (중) '개발' 부서 직원들의 평균 연봉을 계산하세요.
dev = df[df['부서'] == '개발']
print(dev['연봉(만원)'].mean())
```

```python
# 9   (상) 각 부서별로 가장 높은 연봉을 받는 직원의 정보를 출력하세요.
max_sal = df.groupby('부서')['연봉(만원)'].max()
display(df[df['연봉(만원)'].isin(max_sal)])
```

```python
# 10   (상) 입사년도별 직원 수를 계산하고, 그 결과를 입사년도로 정렬하세요.
print(df['입사년도'].value_counts().sort_index())
```

```python
# 11   (상) 각 직급별 연봉의 최대값, 최소값, 평균값을 계산하세요.
print(df.groupby('직급')['연봉(만원)'].max())
print(df.groupby('직급')['연봉(만원)'].min())
print(df.groupby('직급')['연봉(만원)'].mean())
```

```python
# 12   (상) 연봉 상위 10명의 직원 정보를 출력하세요.
ms = df['연봉(만원)'].sort_values(ascending=False).head(10)
display(df[df['연봉(만원)'].isin(ms)].sort_values('연봉(만원)', ascending=False))
```

```python
# 13   (상) 부서를 기준으로 그룹화한 후, 각 그룹 내에서 입사년도가 가장 오래된 직원의 정보를 출력하세요.
longtime = df.groupby('부서')['입사년도'].min()
display(df[df['입사년도'].isin(longtime)].sort_values('부서'))
```

```python
# 14   (상) 각 부서에서 연봉 상위 3명의 직원 정보를 출력하세요.
tp3 = df.groupby('부서')['연봉(만원)'].nlargest(3)
display(df[df['연봉(만원)'].isin(tp3)].sort_values('부서'))
```

```python
# 15   (상) 연봉의 범위(최소값과 최대값 사이의 차이)가 가장 큰 부서를 찾으세요.
rg = (df.groupby('부서')['연봉(만원)'].max() - df.groupby('부서')['연봉(만원)'].min()).nlargest(1)
print(rg)
```
