```python
df['location1'] = df['addr1'].apply(lambda x : x.split(' ')[0])
df
df['location2'] = df['addr1'].apply(lambda x : x.split(' ')[0][:2])
df
```

```python
df_map = df.loc[:, ['cat1', 'cat2', 'cat3', 'location2', 'title', 'mapx', 'mapy']]
df_map.shape
```

```python
# 위경도 object type을 실수인 float64 변환
df_map['mapx'] = df_map['mapx'].astype('float64')
df_map['mapy'] = df_map['mapy'].astype('float64')
```

```python
# 이상치 제거 전 산점도
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_map, x='mapx', y='mapy', s=10)

# x축과 y축의 범위를 설정하여 '줌 인'효과 생성
# plt.xlim(124, 130)
# plt.ylim(33, 39)

plt.show()
```

### 이상치 확인

```python
# 경도의 이상치
df_map['mapx'].describe()
```

```python
# 경도 이상치 제거
qx1 = df_map['mapx'].quantile(.25)
qx3 = df_map['mapx'].quantile(.75)
print(qx1, qx3)

# IQR(interquartile range)
IQR = qx3 - qx1

# 하한값 계산 Q1 - IQR * 1.5
# 상한값 계산 Q3 + IQR * 1.5

# 이상치가 아닌 데이터의 인덱스 찾기
mask_x = (df_map['mapx'] >= (qx1 - IQR * 1.5)) & (df_map['mapx'] <= (qx3 + IQR * 1.5))

# 이상치를 제외한 데이터 필터링
print(len(df_map))
df_map = df_map[mask_x]
print(len(df_map))
```

```python
# 위도 이상치 제거
qy1 = df_map['mapy'].quantile(.25)
qy3 = df_map['mapy'].quantile(.75)
print(qy1, qy3)

# IQR(interquartile range)
IQR = qx3 - qx1

# 하한값 계산 Q1 - IQR * 1.5
# 상한값 계산 Q3 + IQR * 1.5

# 이상치가 아닌 데이터의 인덱스 찾기
mask_y = (df_map['mapy'] >= (qy1 - IQR * 1.5)) & (df_map['mapy'] <= (qy3 + IQR * 1.5))

# 이상치를 제외한 데이터 필터링
print(len(df_map))
df_map = df_map[mask_y]
print(len(df_map))
```

```python
plt.boxplot(df_map['mapx'])
plt.show()
```

```python
# 이상치 제거 후 산점도
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df_map, x='mapx', y='mapy', s=5, hue='location2')

# x축과 y축의 범위를 설정하여 '줌 인'효과 생성
# plt.xlim(124, 130)
# plt.ylim(33, 39)

plt.show()
```
