```python
import requests
import json
import folium
from pyproj import Transformer
```

```python
def convert_coordinates(coordinates):
    transformer = Transformer.from_crs("EPSG:3857", "EPSG:4326")  # EPSG:3857에서 EPSG:4326로 변환
    converted_coordinates = []
    for coord_list in coordinates:
        converted_coords = []
        for coord in coord_list:
            latitude, longitude = transformer.transform(coord[0], coord[1])
            converted_coords.append([latitude, longitude])
        converted_coordinates.append(converted_coords)
    return converted_coordinates

# VWorld API 키
api_key = "DF2F52B0-51A7-3D20-8794-E1F35A69D67E"

# # 지도 초기화
# m = folium.Map(location=[37.5665, 126.9780], zoom_start=12)  # 서울을 중심으로 초기화

# API에서 데이터 가져오기
response = requests.get('https://api.vworld.kr/req/data?key=DF2F52B0-51A7-3D20-8794-E1F35A69D67E&domain=ulsankh.com&service=data&version=2.0&format=json&page=1&data=LT_L_FRSTCLIMB&geometry=true&attribute=true&crs=EPSG:3857&request=getfeature&size=1000&buffer=3000&geomfilter=POINT(13883487.068135787 4576062.330902048)')

# JSON 파일로 저장하기
if response.status_code == 200:
    data = response.json()
    
    # 데이터가 비어있는지 확인
    if data and 'features' in data['response']['result']['featureCollection']:
        features = data['response']['result']['featureCollection']['features']
        
        # 좌표 변환
        for feature in features:
            if 'geometry' in feature and 'coordinates' in feature['geometry']:
                coordinates = feature['geometry']['coordinates']
                feature['geometry']['coordinates'] = convert_coordinates(coordinates)
        
        # 변환된 데이터를 새로운 JSON 파일에 저장
        with open('converted_data.json', 'w', encoding='utf-8') as jsonfile:
            json.dump(data, jsonfile, ensure_ascii=False, indent=4)
    else:
        print("데이터가 비어있습니다.")
else:
    print("API에서 데이터를 가져오지 못했습니다.")

# 좌표 읽기
def extract_coordinates(json_data):
    coordinates = []
    if 'response' in json_data and 'result' in json_data['response'] and 'featureCollection' in json_data['response']['result']:
        features = json_data['response']['result']['featureCollection']['features']
        for feature in features:
            if 'geometry' in feature and 'coordinates' in feature['geometry']:
                coords = feature['geometry']['coordinates']
                coordinates.append(coords)
    return coordinates

# 주어진 JSON 파일 읽기
with open('converted_data.json', 'r', encoding='utf-8') as file:
    json_data = json.load(file)

# 좌표 추출
extracted_coords = extract_coordinates(json_data)

#지도에 출력
for line_coords in extracted_coords:
    folium.PolyLine(locations=line_coords, color='blue').add_to(m)
m
```
