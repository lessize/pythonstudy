```python
# 데이터 분석을 위한 pandas, 수치계산을 위한 numpy
# 시각화를 위한 seaborn, matplotlib.pyplot 을 로드합니다. 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 한글설정 & 마이너스표시
plt.rc('font',family='Malgun Gothic') 
plt.rc('axes',unicode_minus=False)
```

```python
## 모든 열 보이게하기 설정
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('체육활동.csv')
```

```python
df
```

```python
# 컬럼이름
col_names = ['통계분류(1)', '통계분류(2)', '통계분류(3)', '걷기(조깅 속보)', '등산', '체조(맨손 체조 생활 체조)', '볼링', '자전거/사이클/산악자전거','축구/풋살','보디빌딩(헬스)','수영','당구/포켓볼','골프(그라운드 파크 포함)']
col_names
```

```python
# 컬럼명 변경하고 4번 인덱스 행부터 사용하기
# df
df.columns = col_names
df = df[5:27]
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df.loc['걷기(조깅 속보)':'골프(그라운드 파크 포함)']
```

```python
df.shape
```

```python
# 4번째부터 14번째 컬럼의 데이터 타입을 object에서 float으로 변경
for col in df.columns[3:14]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df.dtypes
```

```python
df.loc[0]['걷기(조깅 속보)':'골프(그라운드 파크 포함)'].plot.barh()
plt.show()
df.loc[1]['걷기(조깅 속보)':'골프(그라운드 파크 포함)'].plot.barh()
plt.show()
```
