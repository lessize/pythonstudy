```python
import pandas as pd
import numpy as np
```

### 데이터 셋을 불러와서 처음 5개의 행을 출력하시오.

```python
df = pd.read_csv('data.csv')
df.head(5)
```

### '카테고리' 별로 몇 개의 상품이 있는지 각각 세어 출력하시오.

```python
df['카테고리'].value_counts()
```

### 모든 상품의 평균 가격을 계산하시오.

```python
df['가격'].mean()
```

### '판매수량'이 가장 높은 상위 10개 상품을 출력하시오.

```python
df.nlargest(10, '판매수량')
```

### 각 '카테고리' 별 평균 판매 수량을 계산하시오.

```python
df.groupby('카테고리')['판매수량'].mean()
```

### '판매날짜'를 기준으로 데이터를 정렬하시오. (오래된 순)

```python
df.sort_values(by='판매날짜')
```

### 각 '카테고리' 별로 최고 가격을 가진 상품의 이름을 출력하시오.

```python
df.iloc[df.groupby('카테고리')['가격'].idxmax()]
```

### 모든 상품의 총 판매액(가격 * 판매수량)을 계산하시오.

```python
(df['가격'] * df['판매수량']).sum()
```

### 각 상품별로 '판매액'을 새로운 파생열로 추가하시오.

```python
df['판매액'] = df['가격'] * df['판매수량']
df
```

### 각 '카테고리' 별 총 판매액을 계산하시오.

```python
df.groupby('카테고리')['판매액'].sum()
```

### 각 '카테고리' 별 평균 판매액을 계산하고, 평균 판매액이 가장 높은 카테고리를 출력하시오.

```python
avg = df.groupby('카테고리')['판매액'].mean()
avg
```

```python
avg.idxmax()
```

### 판매 데이터 중에서 '판매수량' 상위 25%에 해당하는 상품들만 선택하여, 그 상품들의 평균 '가격'을 계산하시오.

```python
q = df[df['판매수량'] > df['판매수량'].sort_values(ascending = False).quantile(.75)]
q
```

```python
q['가격'].mean()
```

### '판매날짜'를 datetime 타입으로 변환하고, 연도별 총 판매수량을 계산하시오.

```python
df['판매날짜'] = pd.to_datetime(df['판매날짜'])
df['판매날짜'].info()
```

```python
d_year = df['판매날짜'].dt.year
df.groupby(d_year)['판매수량'].sum()
```

### 각 카테고리 내에서 '가격' 상위 10%에 해당하는 상품들의 목록을 출력하시오.

```python
df[df.groupby('카테고리')['가격'].transform(lambda x: x > x.quantile(0.9))]
```

### '판매날짜'를 기준으로 각 달의 총 판매액을 계산하고, 판매액이 가장 높은 달을 출력하시오.

```python
d_month = df['판매날짜'].dt.month
mxm = df.groupby(d_month)['판매액'].sum()
mxm
```

```python
mxm.idxmax()
```
