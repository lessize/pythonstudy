```python
lat =df['위도'].mean()
lot =df['경도'].mean()
lat,lot
map = folium.Map(location=(lat, lot),zoom_start=10)

# 마커표시
for row in df.index[:len(df.index)]:
    lat = df.loc[row, '위도']
    lot = df.loc[row, '경도']
    
    frtrlNm = df.loc[row, '산명']
    addrNm = df.loc[row,'주소']
    aslAltide = df.loc[row, '해발고도']
    detail_info = f'산명 : {frtrlNm}   / 주소:{addrNm}   / 해발고도:{aslAltide}'
    
    folium.Marker(
        location=[lat,lot],
        tooltip=frtrlNm,
        popup=detail_info,
        icon=folium.Icon(icon="cloud"),
    ).add_to(map)

map
```

```python
df = pd.read_json('mountain_100.json',encoding='UTF-8')
df = pd.json_normalize(df['response']['body']['items']['item'])
df = df.rename(columns={'lot': '경도', 'frtrlId': '아이디','ctpvNm':'시도명','crtrDt':'데이터추출일시','aslAltide':'해발고도','mtnCd':'산코드','addrNm':'주소','frtrlNm':'산명','lat':'위도'})
df.시도명
```

```python
bar_width = 0.15
array_data = df[['시도명']].groupby('시도명').var()
for i in range(len(array_data)):
    df_mountain_array =  df.loc[(df.시도명==array_data.index[i]), ['시도명','해발고도','산명']]   
    display(df_mountain_array)
    name_data = str(df_mountain_array.iloc[0,0])+'[M][bar]'
    x = np.arange(len(df_mountain_array['산명']))

    plt.figure(figsize=(10, 6))


    #---산 해발고도---
    # 각 연도에 대한 막대 그래프를 그립니다.
    plt.bar(x, df_mountain_array['해발고도'], width=bar_width, color='green', label='해발고도')

    plt.title(name_data)  # 그래프 제목 설정
    plt.xlabel('산')  # x 축 레이블 설정
    plt.ylabel('해발고도[M]')  # y 축 레이블 설정
    plt.ylim(0, 1500)
    plt.yticks(np.arange(0, 1501, 1500), labels=[f'{i}M' for i in np.arange(0, 1501, 1500)])  # y 축 눈금 레이블 설정
    plt.xticks(ticks=x, labels=df_mountain_array['산명'], rotation=45)
    plt.legend()  # 범례 표시
    plt.grid(False)  # 격자 표시 안 함
    plt.show()  # 그래프 표시
    
    name_data = str(df_mountain_array.iloc[0,0])+'[M][line]'
    plt.figure(figsize=(10, 6))
    sns.lineplot(data=df_mountain_array, x='산명', y='해발고도', marker='o', color='blue', label='해발고도')

    plt.title(name_data)  # 그래프 제목 설정
    plt.ylim(0, 1500)
    plt.yticks(np.arange(0, 1501, 1500), labels=[f'{i}M' for i in np.arange(0, 1501, 1500)])  # y 축 눈금 레이블 설정
    plt.xticks(ticks=x, labels=df_mountain_array['산명'], rotation=45)
    plt.xlabel('산')  # x 축 레이블 설정
    plt.ylabel('해발고도[M]')  # y 축 레이블 설정
    plt.grid(True)  # 격자 표시
    plt.legend()  # 범례 표시
    plt.show()  # 그래프 표시


```

```python
df = pd.read_json('mountain_100.json',encoding='UTF-8')
df = pd.json_normalize(df['response']['body']['items']['item'])
df = df.rename(columns={'lot': '경도', 'frtrlId': '아이디','ctpvNm':'시도명','crtrDt':'데이터추출일시','aslAltide':'해발고도','mtnCd':'산코드','addrNm':'주소','frtrlNm':'산명','lat':'위도'})
df
```

```python
df.head(10)
bar_width = 0.15
array_data = df[['시도명']].groupby('시도명').var()

df_mountain_array =  df.head(10)
display(df_mountain_array)
name_data = str(df_mountain_array.iloc[0,0])+'[M][bar]'
x = np.arange(len(df_mountain_array['산명']))

plt.figure(figsize=(10, 6))


#---산 해발고도---
# 각 연도에 대한 막대 그래프를 그립니다.
plt.bar(x, df_mountain_array['해발고도'], width=bar_width, color='green', label='해발고도')

plt.title(name_data)  # 그래프 제목 설정
plt.xlabel('산')  # x 축 레이블 설정
plt.ylabel('해발고도[M]')  # y 축 레이블 설정
plt.ylim(0, 1500)
plt.yticks(np.arange(0, 1501, 1500), labels=[f'{i}M' for i in np.arange(0, 1501, 1500)])  # y 축 눈금 레이블 설정
plt.xticks(ticks=x, labels=df_mountain_array['산명'], rotation=45)
plt.legend()  # 범례 표시
plt.grid(False)  # 격자 표시 안 함
plt.show()  # 그래프 표시

name_data = str(df_mountain_array.iloc[0,0])+'[M][line]'
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_mountain_array, x='산명', y='해발고도', marker='o', color='blue', label='해발고도')

plt.title(name_data)  # 그래프 제목 설정
plt.ylim(0, 1500)
plt.yticks(np.arange(0, 1501, 1500), labels=[f'{i}M' for i in np.arange(0, 1501, 1500)])  # y 축 눈금 레이블 설정
plt.xticks(ticks=x, labels=df_mountain_array['산명'], rotation=45)
plt.xlabel('산')  # x 축 레이블 설정
plt.ylabel('해발고도[M]')  # y 축 레이블 설정
plt.grid(True)  # 격자 표시
plt.legend()  # 범례 표시
plt.show()  # 그래프 표시



df_mountain_array =  df.tail(10)
display(df_mountain_array)
name_data = str(df_mountain_array.iloc[0,0])+'[M][bar]'
x = np.arange(len(df_mountain_array['산명']))

plt.figure(figsize=(10, 6))


#---산 해발고도---
# 각 연도에 대한 막대 그래프를 그립니다.
plt.bar(x, df_mountain_array['해발고도'], width=bar_width, color='green', label='해발고도')

plt.title(name_data)  # 그래프 제목 설정
plt.xlabel('산')  # x 축 레이블 설정
plt.ylabel('해발고도[M]')  # y 축 레이블 설정
plt.ylim(0, 1500)
plt.yticks(np.arange(0, 1501, 1500), labels=[f'{i}M' for i in np.arange(0, 1501, 1500)])  # y 축 눈금 레이블 설정
plt.xticks(ticks=x, labels=df_mountain_array['산명'], rotation=45)
plt.legend()  # 범례 표시
plt.grid(False)  # 격자 표시 안 함
plt.show()  # 그래프 표시

name_data = str(df_mountain_array.iloc[0,0])+'[M][line]'
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_mountain_array, x='산명', y='해발고도', marker='o', color='blue', label='해발고도')

plt.title(name_data)  # 그래프 제목 설정
plt.ylim(0, 1500)
plt.yticks(np.arange(0, 1501, 1500), labels=[f'{i}M' for i in np.arange(0, 1501, 1500)])  # y 축 눈금 레이블 설정
plt.xticks(ticks=x, labels=df_mountain_array['산명'], rotation=45)
plt.xlabel('산')  # x 축 레이블 설정
plt.ylabel('해발고도[M]')  # y 축 레이블 설정
plt.grid(True)  # 격자 표시
plt.legend()  # 범례 표시
```
