```python
# 데이터 분석을 위한 pandas, 수치계산을 위한 numpy
# 시각화를 위한 seaborn, matplotlib.pyplot 을 로드합니다. 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc

font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('연간_탐방객_수_20240321173335.csv', encoding='cp949')
df = df.iloc[1:,:]
df['총원 평균'] = (df['2020'] + df['2021'] + df['2022'])/3
df
```

```python
bar_width = 0.15

# x 좌표 설정
x = np.arange(len(df['공원별']))

plt.figure(figsize=(20, 6))

# 각 연도에 대한 막대 그래프를 그립니다.
plt.bar(x - bar_width, df['2020'], width=bar_width, color='blue', label='2020')
plt.bar(x, df['2021'], width=bar_width, color='green', label='2021')
plt.bar(x + bar_width, df['2022'], width=bar_width, color='red', label='2022')
plt.bar(x + bar_width+ bar_width, df['총원 평균'], width=bar_width, color='lime', label='총원 평균')

plt.title('연간 탐방객 수[bar]')  # 그래프 제목 설정
plt.xlabel('공원별')  # x 축 레이블 설정
plt.ylabel('탐방객 수')  # y 축 레이블 설정
plt.ylim(0, 10000000)
plt.yticks(np.arange(0, 10000001, 1000000), labels=[f'{i}명' for i in np.arange(0, 10000001, 1000000)])  # y 축 눈금 레이블 설정
plt.xticks(x, df['공원별'])  # x 축 눈금 레이블 설정
plt.legend()  # 범례 표시
plt.grid(False)  # 격자 표시 안 함
plt.show()  # 그래프 표시

plt.figure(figsize=(20, 6))
sns.lineplot(data=df, x='공원별', y='2020', marker='o', color='blue', label='2020')
sns.lineplot(data=df, x='공원별', y='2021', marker='o', color='green', label='2021')
sns.lineplot(data=df, x='공원별', y='2022', marker='o', color='red', label='2022')
sns.lineplot(data=df, x='공원별', y='총원 평균', marker='o', color='lime', label='총원 평균')

plt.title('연간 탐방객 수[line]')  # 그래프 제목 설정
plt.ylim(0, 10000000)
plt.yticks(np.arange(0, 10000001, 1000000), labels=[f'{i}명' for i in np.arange(0, 10000001, 1000000)])  # y 축 눈금 레이블 설정
plt.xlabel('산')  # x 축 레이블 설정
plt.ylabel('탐방객 수')  # y 축 레이블 설정
plt.grid(True)  # 격자 표시
plt.legend()  # 범례 표시
plt.show()  # 그래프 표시
```

```python
df2 = pd.read_csv('여가시간_주요활동_1순위__20240322161433.csv')
df2
```

```python
df_age = df2.iloc[6:15,:]
df_age = df_age[df_age['응답자특성별(3)'].str.endswith('소계')]
df_age[['응답자특성별(2)','2022','2022.1','2022.2','2022.3','2022.4','2022.5','2022.6','2022.7']]
df_mu = df2[df2['응답자특성별(2)'].str.endswith('응답자특성별(2)')]

df_age_s = df_age[df_age['응답자특성별(2)'].str.endswith('청소년(만10~19세)')]
df_age_m = df_age[df_age['응답자특성별(2)'].str.endswith('성인(만20~59세)')]
df_age_b = df_age[df_age['응답자특성별(2)'].str.endswith('60대(만60~69세)')]
df_mu = df_mu.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='활동')
df_mu = df_mu.iloc[2:,1:]
df_age_s = df_age_s.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='청소년(만10~19세)')
df_age_s = df_age_s.iloc[2:,1:]
df_age_s['청소년(만10~19세)'] = df_age_s['청소년(만10~19세)'].astype(float)
df_age_m = df_age_m.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='성인(만20~59세)')
df_age_m = df_age_m.iloc[2:,1:]
df_age_m['성인(만20~59세)'] = df_age_m['성인(만20~59세)'].astype(float)
df_age_b = df_age_b.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='60대(만60~69세)')
df_age_b = df_age_b.iloc[2:,1:]
df_age_b['60대(만60~69세)'] = df_age_b['60대(만60~69세)'].astype(float)
df_age_result = pd.merge(df_mu,df_age_s)
df_age_result = pd.merge(df_age_result,df_age_m)
df_age_result = pd.merge(df_age_result,df_age_b)
df_age_result = df_age_result.iloc[:,1:]
df_age_result
```

```python
df_gender = df2.iloc[15:20,:]
df_gender = df_gender[df_gender['응답자특성별(3)'].str.endswith('소계')]
df_gender[['응답자특성별(2)','2022','2022.1','2022.2','2022.3','2022.4','2022.5','2022.6','2022.7']]
df_gender_m = df_gender[df_gender['응답자특성별(2)'].str.endswith('남성')]
df_gender_s = df_gender[df_gender['응답자특성별(2)'].str.endswith('여성')]
df_gender_m = df_gender_m.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='남성')
df_gender_m = df_gender_m.iloc[2:,1:]
df_gender_m['남성'] = df_gender_m['남성'].astype(float)
df_gender_s = df_gender_s.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='여성')
df_gender_s = df_gender_s.iloc[2:,1:]
df_gender_s['여성'] = df_gender_s['여성'].astype(float)
df_gender_result = pd.merge(df_mu,df_gender_m)
df_gender_result = pd.merge(df_gender_result,df_gender_s)
df_gender_result = df_gender_result.iloc[:,1:]
df_gender_result
```

```python
df_array = df2.iloc[33:,:]
df_array = df_array[df_array['응답자특성별(3)'].str.endswith('소계')]
df_array[['응답자특성별(2)','2022','2022.1','2022.2','2022.3','2022.4','2022.5','2022.6','2022.7']]
df_array_b = df_array[df_array['응답자특성별(2)'].str.endswith('대도시')]
df_array_m = df_array[df_array['응답자특성별(2)'].str.endswith('중소도시')]
df_array_s = df_array[df_array['응답자특성별(2)'].str.endswith('읍/면지역')]
df_array_b = df_array_b.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='대도시')
df_array_b = df_array_b.iloc[2:,1:]
df_array_b['대도시'] = df_array_b['대도시'].astype(float)
df_array_m = df_array_m.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='중소도시')
df_array_m = df_array_m.iloc[2:,1:]
df_array_m['중소도시'] = df_array_m['중소도시'].astype(float)
df_array_s = df_array_s.melt(id_vars='응답자특성별(2)', var_name='년도', value_name='읍/면지역')
df_array_s = df_array_s.iloc[2:,1:]
df_array_s['읍/면지역'] = df_array_s['읍/면지역'].astype(float)
df_array_result = pd.merge(df_mu,df_array_b)
df_array_result = pd.merge(df_array_result,df_array_m)
df_array_result = pd.merge(df_array_result,df_array_s)
df_array_result = df_array_result.iloc[:,1:]
df_array_result
```
