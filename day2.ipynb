```python
def calculate(x, y) :
  add = x + y
  difference = x - y
  multifly = x * y
  mod = x // y
  remain = x % y
  return add, difference, multifly, mod, remain

print(sum(10, 30))

print(type(calculate(10, 3)))

add, difference, multifly, mod, remain = calculate(10, 3)
print(add, difference, multifly, mod, remain )
```

```python
# 1~100까지 정수 중에 짝수의 합과 홀수의 합을 각각 구하시오
sumOfEven = 0
sumOfOdd = 0

for num in range(1, 101) :
  if (num%2 == 0) :
    sumOfEven += num
  else :
    sumOfOdd += num

print(f'sumOfEven={sumOfEven}')
print(f'sumOfOdd={sumOfOdd}')
```

```python
# 구구단 출력하기 2~9단까지 중첩 for문 이용, f'' 사용
for dan in range(2, 9+1):
  print(f'=={dan}단 출력==')
  for num in range(1, 9+1):
    print(f'{dan:2} * {num:2}={dan*num:2}')
```

```python
# 단수를 입력받아 해당 구구단을 출력하시오

dansu = int(input('단수를 입력하세요 : '));
# print(type(dansu))

for num in range(1, 9+1):
  print(f'{dansu} * {num} = {dansu * num:>2} ')
# > : 오른쪽 정렬, < : 왼쪽 정렬, ^ : 가운데 정렬, :2 : 두자릿수 자리 정렬
```

```python
def sum_all(*args):
  print(type(args))

  for ele in args :
    print(ele)


sum_all(1, 2, 3)
```

```python
# 가변인자 *:튜플, **:딕셔너리
def sum_all2(*args1, **args2):
  print(type(args1))
  print(type(args2))

  for ele in args1 :
    print(ele)

  for key, val in args2.items():
    print(key, val)

# case1)
sum_all2(1, 2, 3, name='홍길동', age=10, gender='남자')

# case2)
x = (1, 2, 3)
y = {'name':'홍길동', 'age':10, 'gender':'남자'}
sum_all2(*x, **y)
```
