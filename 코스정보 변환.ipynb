```python
import csv
import json

# csv 파일 경로
csv_file_path = '코스정보.csv'

# csv 파일 읽어오기
with open(csv_file_path, 'r', encoding='cp949') as f:
    reader = csv.reader(f)
    next(reader)  # 첫 줄 skip

    # 각 라인마다 딕셔너리 생성 후 리스트에 추가
    data = []
    for line in reader:
        d = {
              "COURS_IDNO": int(line[0]),
              "COURS_DATA_SET_TPCD": line[1],
              "COURS_NM": line[2],
              "COURS_SNNM": line[3],
              "COURS_DSCRT": line[4],
              "COURS_CTPRV_NM": line[5],
              "COURS_EMNDN_NM": line[6],
              "COURS_LTNO": line[7],
              "COURS_ARA_NM": line[8],
              "COURS_WEBS_ADDR": line[9],
              "COURS_TOT_LNGTH": float(line[10]),
              "COURS_AVRG_GRDNT": float(line[11]),
              "COURS_TOP_HASLV": float(line[12]),
              "COURS_LOWST_HASLV": float(line[13]),
              "COURS_UPR_LNGTH": float(line[14]),
              "COURS_FTL_LNGTH": float(line[15]),
              "COURS_DST_LNGTH": float(line[16]),
              "COURS_DGDFF_VAL": int(line[17]),
              "DATA_STCD": line[18],
              "COURS_RQRMN_TM": int(line[19]),
              "DATA_RGSTN_DTM": line[20],
              "DATA_UPDT_DTM": line[21],
              "DATA_CONNO": int(line[22]),
              "OPERT_TPCD": line[23],
              "OCCRR_DTM": int(line[24]),
              "CNTCT_BGN_DTM": int(line[25]),
              "CNTCT_PRCES_DTM": int(line[26]),
              "CNTCT_PRCES_STCD": line[27],
              "TRNSA_ID": line[28]
        }
        data.append(d)

# json string으로 변환
json_string = json.dumps(data, ensure_ascii=False, indent=2)

print(json_string)

# txt 파일로 저장할 경로
txt_file_path = '코스정보.json'

# txt 파일 쓰기
with open(txt_file_path, 'w', encoding='utf-8') as f:
    f.write(json_string)
```
